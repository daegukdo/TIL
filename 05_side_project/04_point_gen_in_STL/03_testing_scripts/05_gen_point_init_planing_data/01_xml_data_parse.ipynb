```python
import numpy as np
from xml.etree.ElementTree import parse

path_xml_data = 'raw_data/'
path_ept = 'data/'

file_ext = '.xml'
xml_data_name = ['c1L_plan', 
                 'c1R_plan', 
                 'c2L_plan', 
                 'c2R_plan']

FT_prefix = ['F', 'T']
```

```python
for fi in range(len(xml_data_name)):
    tree = parse(path_xml_data + xml_data_name[fi] + file_ext)
    main_root = tree.getroot()

    femur_regi = main_root.getchildren()[2].getchildren()[2].getchildren()
    tibia_regi = main_root.getchildren()[3].getchildren()[2].getchildren()

    femur_regi_data = []
    for i in range(len(femur_regi)):
        value = femur_regi[i].find("target").text
        value_list = value.split(',')
        tmp = []
        for j in range(len(value_list)):
            tmp.append(float(value_list[j]))
        femur_regi_data.append(tmp)
        
    np.save(path_ept + FT_prefix[0] + xml_data_name[fi], np.array(femur_regi_data))

    tibia_regi_data = []
    for i in range(len(tibia_regi)):
        value = tibia_regi[i].find("target").text
        value_list = value.split(',')
        tmp = []
        for j in range(len(value_list)):
            tmp.append(float(value_list[j]))
        tibia_regi_data.append(tmp)
        
    np.save(path_ept + FT_prefix[1] + xml_data_name[fi], np.array(tibia_regi_data))
```
