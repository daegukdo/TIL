### step별 데이터 확인 및 검증

 - 각 step 진행에 따른 검증이 필요하므로 이를 point 데이터로 저장하고 다른 상용툴로 확인 진행

```python
import numpy as np

# 다시 읽어드리기

points40_with_error_on_plane = np.load('data/points40_with_error_on_plane.npy')
nearest_points40_with_error_on_plane = np.load('data/nearest_points40_with_error_on_plane.npy')
points40_with_uncertainty_error = np.load('data/points40_with_uncertainty_error.npy')
```

```python
print(len(points40_with_error_on_plane))
print(len(nearest_points40_with_error_on_plane))
print(len(points40_with_uncertainty_error))
```

```python
file1 = open("data/01r_points40_with_error_on_plane.txt","w") 

for i in range(len(points40_with_error_on_plane)):
    tmp_str = ""
    for j in range(3):
        tmp_str = tmp_str + str(points40_with_error_on_plane[i][j])
        if j != 2:
            tmp_str = tmp_str + ", "
        else:
            tmp_str = tmp_str + "\n"
    
    file1.write(tmp_str)
            
file1.close()
```

```python
file2 = open("data/02r_nearest_points40_with_error_on_plane.txt","w") 

for i in range(len(nearest_points40_with_error_on_plane)):
    tmp_str = ""
    for j in range(3):
        tmp_str = tmp_str + str(nearest_points40_with_error_on_plane[i][j])
        if j != 2:
            tmp_str = tmp_str + ", "
        else:
            tmp_str = tmp_str + "\n"
    
    file2.write(tmp_str)
            
file2.close()
```

```python
file3 = open("data/03r_points40_with_uncertainty_error.txt","w") 

for i in range(len(points40_with_uncertainty_error)):
    tmp_str = ""
    for j in range(3):
        tmp_str = tmp_str + str(points40_with_uncertainty_error[i][j])
        if j != 2:
            tmp_str = tmp_str + ", "
        else:
            tmp_str = tmp_str + "\n"
    
    file3.write(tmp_str)
            
file3.close()
```
