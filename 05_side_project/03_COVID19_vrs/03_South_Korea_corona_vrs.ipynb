```python
# draw charts
%matplotlib notebook

from matplotlib import pyplot as plt
import numpy as np
```

```python
# CSV 파일 읽어들이기

import csv  # csv 모듈 임포트
import pprint

# 파일 위치, 경로 정의
file_path = '01_South_Korea/route.csv'

# route.csv 파일 열기
with open(file_path) as file:
    reader = csv.reader(file)  # CSV 파일을 읽어들이는 읽기 객체
    raw = list(reader)      #  CSV 파일 내용을 리스트로 읽어들인다

pprint.pprint(raw[:10])  # 읽어들인 내용을 화면에 출력
```

```python
# 안내를 위한 row를 삭제
raw_ = raw[1:]
```

```python
province_of_patient = []
city_of_patient = []
visit_of_patient = []
latitude_city = []
longitude_city = []

for i in range(len(raw_)):
    province_name = raw_[i][2]
    city_name = raw_[i][3]
    visit_name = raw_[i][4]
    latitude_value = raw_[i][5]
    longitude_value = raw_[i][6]
    
    province_of_patient.append(province_name)
    city_of_patient.append(city_name)
    visit_of_patient.append(visit_name)
    latitude_city.append(latitude_value)
    longitude_city.append(longitude_value)
```

```python
import folium
southkorea_map = folium.Map(location=[36.55,126.983333], zoom_start=7, titles='loation of patient in South Korea')
```

```python
for lat, lon, province, city, visit in zip(latitude_city, longitude_city, province_of_patient, city_of_patient, visit_of_patient):
    popup_data = folium.map.Popup(html = ('<b>' + 'Province: ' + '</b>' + str(province) + '<br>' + 
                                          '<b>' + 'City: ' + '</b>' + str(city) + '<br>' + 
                                          '<b>' + 'Visit: ' + '</b>' + str(visit) + '<br>'),
                                  max_width = 250)
    folium.CircleMarker([lat, lon],
                        radius=5,
                        color='red',
                        popup =popup_data,
                        fill_color='red',
                        fill_opacity=0.5 ).add_to(southkorea_map)
```
