## 1. dicom image historgram equalization

 - sawbone CT 데이터의 intensity가 잘 구별되지 않음
 
 - 이를 위해 image historgram equalization을 통해 intensity를 살려봄 

 - 참고 자료
 
 > https://github.com/paultpearson/DSB16/blob/master/data/tutorial_02_plotting_and_histogram_equalization.py

 - data를 display하여 확인

```python
# import module
import pydicom
import numpy as np
import os
import matplotlib.pyplot as plt

# set jupyter property(?)
%matplotlib inline
```

```python
# read sawbone.dcm
filepath = os.path.join("dcm_data", "sawbone.dcm")

dc = pydicom.dcmread(filepath)
```

```python
dc_pxd = dc.pixel_array

fig1  = plt.figure()
plt.imshow(dc_pxd, cmap="gray", interpolation="bicubic")
plt.colorbar()
fig1.suptitle("sawbone CT data with gray colormap", fontsize=12)
```

 - histogram equalization은 intensity 분포를 histogram 상에서 조절하여 이미지 전체의 contrast를 조절하는 방법
 
 - 참고자료
 
 > https://www.math.uci.edu/icamp/courses/math77c/demos/hist_eq.pdf

 - dicom image의 histogram 보기
 
 - hist 함수를 사용

```python
plt.hist(dc_pxd)
```

 - histogram equalization을 진행
 
 - scikit-image module의 exposure 함수를 사용
 
 - 다시 histogram을 확인해보자

```python
# import module
from skimage import exposure
```

 - 여기서 x-axis의 범위가 0~1로 바뀐 것과 전반적인 분포를 띄는 것을 볼 수 있음

```python
dc_pxd_hsteq = exposure.equalize_hist(dc_pxd)
plt.hist(dc_pxd_hsteq)
```

 - histogra equalization을 진행한 pixel data(ndarray)를 이미지로 확인해보자

```python
fig2  = plt.figure()
plt.imshow(dc_pxd_hsteq, cmap="gray", interpolation="bicubic")
plt.colorbar()
fig2.suptitle("hist. equalization sawbone CT data with gray colormap", fontsize=12)
```

 ## 2. dicom image processing - intensity enhancing
 
  - 먼저 sawbone의 intensity만 enhance하기 위해서 sawbone intensity 대역을 확인해야 함
  
  - 굉장히 비정성적인 방법으로 imageJ를 사용하여 sawbone intensity를 확인하였음 ... 약 300~500 정도의 값을 가짐
  
  - 조건문을 사용하여 이 부분에만 intensity 값을 올려줌
  
  - 위의 dc 데이터를 그대로 사용
  
  - 참고자료
  
  > https://imagej.nih.gov/ij/download.html

```python
for n,val in enumerate(dc.pixel_array.flat):
    if (val>300 and val<500):
        dc.pixel_array.flat[n]=dc.pixel_array.flat[n]+1500
        
dc.PixelData = dc.pixel_array.tobytes()
```

 - histogram을 그려 보면 특정 부분만 값이 증가한 것을 확인할 수 있음

```python
dc_pxd_enhanced = dc.pixel_array

plt.hist(dc_pxd_enhanced)
```

 - image를 display 해보면 bone 부분만 강조된 것을 확인할 수 있음

```python
fig3  = plt.figure()
plt.imshow(dc_pxd_enhanced, cmap="gray", interpolation="bicubic")
plt.colorbar()
fig3.suptitle("enhanced sawbone CT data with gray colormap", fontsize=12)
```
