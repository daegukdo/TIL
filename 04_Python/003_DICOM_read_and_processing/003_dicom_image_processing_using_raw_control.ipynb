### data 읽어오기

 - dcm 확장자의 데이터를 읽기

```python
import numpy as np
import pydicom
import glob
import os

dcmlists = glob.glob(os.path.join("dcm_data_v2", "*.dcm"))
```

### pixel data 값을 변화

 - contrast를 변경하고자하는 pixel 영역을 선택
 
 - 이후 이 영역 내의 pixel 값을 조정 (값을 증가)
 
 - 선택한 pixel들의 값이 올라간 것을 확인할 수 있음

```python
for i in range(len(dcmlists)):
    file_path = dcmlists[i]
    file_name = os.path.basename(file_path)
    
    dc = pydicom.dcmread(file_path)
    
    for n,val in enumerate(dc.pixel_array.flat):
        # pixel 값 영역 선택
        if (val>2250 and val<2650):
            # pixel 값을 조정
            dc.pixel_array.flat[n]=dc.pixel_array.flat[n]+2048
        
    dc.PixelData = dc.pixel_array.tobytes()
    
    result_file_name = os.path.join("dcm_data_v2_cvt", file_name)
    
    dc.save_as(result_file_name)
    
    print(result_file_name + " --> fin.")
```

### 값이 조정되었는지 비교해보기

 - pixel 값 조정 전의 데이터를 읽기
 
 - pixel 값 조정 후의 데이터를 읽기
 
 - 두 dcm 데이터를 디스플레이하여 확인해보기

```python
import matplotlib.pyplot as plt

# 데이터 디스플레이 함수
def sample_stack(stack, rows=3, cols=3, start_with=0, show_every=1):
    fig,ax = plt.subplots(rows,cols,figsize=[8,8])
    for i in range(rows*cols):
        ind = start_with + i*show_every
        ax[int(i/rows),int(i % rows)].set_title('slice %d' % ind)
        ax[int(i/rows),int(i % rows)].imshow(stack[ind],cmap='gray')
        ax[int(i/rows),int(i % rows)].axis('off')
    plt.show()

# 데이터를 load하는 함수들
def load_scan(path):
    slices = [pydicom.read_file(path + '/' + s) for s in os.listdir(path)]
    slices.sort(key = lambda x: int(x.InstanceNumber))
    try:
        slice_thickness = np.abs(slices[0].ImagePositionPatient[2] - slices[1].ImagePositionPatient[2])
    except:
        slice_thickness = np.abs(slices[0].SliceLocation - slices[1].SliceLocation)
        
    for s in slices:
        s.SliceThickness = slice_thickness
        
    return slices

def get_pixels_hu(scans):
    image = np.stack([s.pixel_array for s in scans])
    # Convert to int16 (from sometimes int16), 
    # should be possible as values should always be low enough (<32k)
    image = image.astype(np.int16)
    
    # Convert to Hounsfield units (HU)
    intercept = scans[0].RescaleIntercept
    slope = scans[0].RescaleSlope
    
    if slope != 1:
        image = slope * image.astype(np.float64)
        image = image.astype(np.int16)
        
    image += np.int16(intercept)
    
    return np.array(image, dtype=np.int16)
```

```python
dcm_data = load_scan('dcm_data_v2')
dcm_data_cvt = load_scan('dcm_data_v2_cvt')
```

```python
img = get_pixels_hu(dcm_data)
img_cvt = get_pixels_hu(dcm_data_cvt)
```

```python
sample_stack(img)
```

```python
sample_stack(img_cvt)
```
