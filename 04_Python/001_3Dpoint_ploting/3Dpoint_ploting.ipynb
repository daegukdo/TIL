```python
%matplotlib notebook

import pandas as pd

# This import registers the 3D projection, but is otherwise unused.
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import

import matplotlib.pyplot as plt
```

#### 3차원 point 데이터 읽기

 - 3차원 점 데이터는 ,(콤마)를 구분자로 가지는 csv로 읽어드림
 - 이때 pandas(판다스)를 사용하여 데이터를 list 형식으로 저장해둠
 - 참고로 이 csv 데이터는 transforamtion matrix --> 따라서 translation term만 읽어서 display에 사용

```python
gooddata_files = "good.txt"
baddata_files = "bad.txt"
```

```python
gooddata_df = pd.read_csv(gooddata_files, header=None)
baddata_df = pd.read_csv(baddata_files, header=None)
```

```python
gooddata = gooddata_df.values
baddata = baddata_df.values
```

#### 3차원 공간에 point scattering 하기

 - 위에서 변환한 3차원 점 데이터 list를 사용
 - for loop를 사용하여 하나의 점씩 display
 - 좋은 point들과 나쁜 point를 붉은색 세모, 파란색 동그라미로 표기
 - 디스플레이는 matplotlib의 pyplot을 사용
 - 위에서 변환하여 받은 list 자료형 중 12,13,14에 해당하는 부분이 각각 x,y,z의 translation term

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for i in range(len(gooddata)):
    xs = gooddata[i][12]
    ys = gooddata[i][13]
    zs = gooddata[i][14]
    ax.scatter(xs, ys, zs, c='blue', marker='o')

for i in range(len(baddata)):
    xs = baddata[i][12]
    ys = baddata[i][13]
    zs = baddata[i][14]
    ax.scatter(xs, ys, zs, c='red', marker='^')
    
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for i in range(int(len(gooddata)/10)):
    xs = gooddata[i*10][12]
    ys = gooddata[i*10][13]
    zs = gooddata[i*10][14]
    ax.scatter(xs, ys, zs, c='blue', marker='o')

for i in range(int(len(baddata)/10)):
    xs = baddata[i*10][12]
    ys = baddata[i*10][13]
    zs = baddata[i*10][14]
    ax.scatter(xs, ys, zs, c='red', marker='^')
    
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
