### binary로 저장된 STL file 읽기

 - STL file은 ASCII 방식 그리고 binary 방식으로 저장될 수 있음
 
 - ASCII는 일반 파일에서 line을 읽어서 처리하듯이 진행할 수 있음
 
 - bianry는 특정 byte씩 나누어서 읽어 정보를 나누어야 함
 
 > ref : https://www.fabbers.com/tech/STL_Format

```python
# set STL file path
file_path = "data/CV.stl"
```

```python
import struct

# read STL file using open() with 'rb' keyword (rb : read binary)
with open(file_path, "rb") as parse :
    
    # first 80 bytes : header
    head = parse.read(80)
    
    # 4 bytes : number of facets in file
    num_of_triangles_raw = parse.read(4)
    # 4 bytes to unsigned long int
    num_of_triangles=struct.unpack("<L", num_of_triangles_raw)
    print("number of facets : " + str(num_of_triangles))
    print("   ")
    
    # read 5 facets
    data_str = ""
    for i in range(5):
        print("facet number : " + str(i))
        
        # 1) 4 * 3 = normal (i, j, k)
        data_str = ""
        for ii in range(3):
            linedata = parse.read(4)
            # 4 bytes to float
            linedata = struct.unpack("<f", linedata)
            data_str = data_str + str(linedata)
        print(data_str)
            
        # 2) 4 * 3 = vertex 1 (x, y, z)
        data_str = ""
        for ii in range(3):
            linedata = parse.read(4)
            # 4 bytes to float
            linedata = struct.unpack("<f", linedata)
            data_str = data_str + str(linedata)
        print(data_str)
            
        # 3) 4 * 3 = vertex 2 (x, y, z)
        data_str = ""
        for ii in range(3):
            linedata = parse.read(4)
            # 4 bytes to float
            linedata = struct.unpack("<f", linedata)
            data_str = data_str + str(linedata)
        print(data_str)
            
        # 4) 4 * 3 = vertex 3 (x, y, z)
        data_str = ""
        for ii in range(3):
            linedata = parse.read(4)
            # 4 bytes to float
            linedata = struct.unpack("<f", linedata)
            data_str = data_str + str(linedata)
        print(data_str)
            
        # attribute
        for iii in range(1):
            linedata=parse.read(2)
            # 2 bytes to unsigned int
            linedata=struct.unpack("<H", linedata)
        print("   ")
```
